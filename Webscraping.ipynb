```python
# import important library

import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# Fetc
URL = "https://realpython.github.io/fake-jobs/"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
```

```python
soup
```

```python
results=soup.find(id='ResultsContainer')
```

```python
results.prettify()
job_elements=results.find_all('div', class_="card-content")
```

```python
job_elements
```

```python
list_title=[]
list_company=[]
list_location=[]
list_posting_date=[]

for job_element in job_elements:
  title=job_element.find('h2',class_="title").text.strip()
  company=job_element.find('h3',class_="subtitle").text.strip()
  location=job_element.find('p',class_="location").text.strip()
  posting_date=job_element.find('time').text.strip()
  list_title.append(title)
  list_company.append(company)
  list_location.append(location)
  list_posting_date.append(posting_date)
```

```python
df=pd.DataFrame({"Job Title":list_title,"Company Name":list_company,"Location":list_location,"Posting Date":list_posting_date})
```

```python
df
```
